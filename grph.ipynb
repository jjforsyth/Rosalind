Returns an adjacency list (of base 3 overlap) from a list of 10 DNA strings.

See problem here: https://rosalind.info/problems/grph/

```python
# Opens file and reads it
with open('./Text_Files/rosalind_grph.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n', '')
```

```python
# Head of file
lines[0:9]
```

```python
# Full file
lines
```

```python
# Combines substrings into strings (formatting from rosalind download) and adds them to a list of fasta sequences
dna_strings=[]
num=0
full_dna=""
for l in range(len(lines)):
    if lines[l].startswith('>'):
        pass
    elif l == len(lines)-1:
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        break
    elif lines[l+1].startswith('>'):
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        full_dna=""
    else:
        if full_dna=="":
            full_dna=lines[l]
        else:
            full_dna=full_dna+lines[l]
```

```python
# Print dna strings
dna_strings
```

```python
# List of fasta names
fasta_names=[]
for l in lines:
    if l.startswith('>'):
        fasta_names.append(l[1:])
```

```python
# Print fasta names
fasta_names
```

```python
# Makes lists of what's has 3 window overlaps, and puts them next to eachother in a list (both names, then sequences to help verify)
adj_list=[]
fasta_adj_list=[]
n=3
for i in range(len(dna_strings)):
    for j in range(i+1, len(dna_strings)):
        a_s = dna_strings[i][:n]
        a_e = dna_strings[i][len(dna_strings[i])-n:]
        b_s = dna_strings[j][:n]
        b_e = dna_strings[j][len(dna_strings[j])-n:]

        if a_s == b_e:
            adj_list.append(dna_strings[j])
            adj_list.append(dna_strings[i])
            fasta_adj_list.append(fasta_names[j])
            fasta_adj_list.append(fasta_names[i])
        if a_e == b_s:
            adj_list.append(dna_strings[i])
            adj_list.append(dna_strings[j])
            fasta_adj_list.append(fasta_names[i])
            fasta_adj_list.append(fasta_names[j])
        else:
            pass
```

```python
# Combines the adjacent sequences in the list
final_adj_list=[]
for i in range(0,len(fasta_adj_list), 2):
    final_adj_list.append(fasta_adj_list[i]+ " "+fasta_adj_list[i+1])
```

```python
# Prints like this for rosalind format requirements.
print('\n'.join(final_adj_list))
```
