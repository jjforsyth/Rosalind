Given: A positive integer n (nâ‰¤1000) and an adjacency list corresponding to a graph on n nodes that contains no cycles.

Return: The minimum number of edges that can be added to the graph to produce a tree.

See more here: https://rosalind.info/problems/tree/

```python
# Opens file and reads it
with open('./Text_Files/rosalind_tree.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n', '')
```

```python
all_nodes=range(1,int(lines[0])+1) # Some nodes may be missing from the list, used to add to the tree in
```

```python
#Converting the lines to sets and integers for simplificaiton.
set_lines = []
for line in lines[1:]:
    set_lines.append(set(
        int(i) for i in line.split() #Weird bit of code required -- splitting the lines, but converting the strings to integers while I do it. 
    ))
```

```python
set_lines
```

```python
#This essentially makes connected 'branches' of the 'tree'.
def combine_sets(sets):
    combined = []  # Initialize an empty list to store the final combined sets
    while sets:  # Continue as long as there are sets to process
        first, *rest = sets  # Split the first set from the rest of the sets
        lf = -1  # Initialize a variable to keep track of the length of the 'first' set
        while len(first) > lf:  # Continue as long as the size of 'first' is increasing
            lf = len(first)  # Update 'lf' to the current size of 'first'
            rest2 = []  # Initialize an empty list to store sets that do not overlap with 'first'
            for r in rest:  # Iterate over the remaining sets
                if first.intersection(r):  # Check if there is any overlap between 'first' and the current set 'r'
                    first |= r  # If there is an overlap, merge 'r' into 'first' using the union operator
                else:
                    rest2.append(r)  # If there is no overlap, add 'r' to 'rest2'
            rest = rest2  # Update 'rest' to 'rest2', which contains only the sets that did not overlap with 'first'
        combined.append(first)  # Add the merged 'first' set to 'combined'
        sets = rest  # Update 'sets' to 'rest', which contains the remaining sets to be processed
    return combined  
```

```python
sets_list = combine_sets(set_lines)
print(sets_list)
print("Length: ", len(sets_list))
```

```python
all_nodes=range(1,int(lines[0])+1) # Some nodes may be missing from the list

unfound = []
for node in all_nodes:
    found = any(node in s for s in sets_list)
    if found: # If it's in the list, we don't want it
        pass
    else: # If it's not in the list, we want to add it as individual nodes to the graph.
        unfound.append(node)

unfound
```

```python
final_answer = (len(sets_list)-1)+len(unfound) #sets_list - 1 because it takes 1 edge to connect 2 nodes
print(final_answer)
```
