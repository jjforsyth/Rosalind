Given: A collection of at most 10 symbols defining an ordered alphabet, and a positive integer n (nâ‰¤10).

Return: All strings of length n that can be formed from the alphabet, ordered lexicographically (use the standard order of symbols in the English alphabet).

See more info here: https://rosalind.info/problems/lexf/

```python
# Opens file and reads it
with open('./Text_Files/rosalind_lexf.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n', '')

lines
```

```python
s_l = [i for i in lines[0].split()]
s_n = int(lines[1])
print(s_l)
print(s_n)
```

```python
from itertools import product

def lexf(s_l, s_n):
    p = list(product(s_l, repeat=s_n))  # Gets all combinations with repetition
    joined_p = [''.join(i) for i in p]  # Combines them into 1 string

    return sorted(joined_p)  # Returns a sorted list

final_answer = lexf(s_l, s_n)
for i in final_answer:  # Done this way to enter into rosalind prompt.
    print(i)
```
