Finds the longest common substring from a collection FASTA sequences.

See problem here: https://rosalind.info/problems/lcsm/

```python
from difflib import SequenceMatcher
```

```python
with open('./Text_Files/rosalind_lcsm.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n', '')
```

```python
lines
```

```python
dna_strings=[]
num=0
full_dna=""
for l in range(len(lines)):
    if lines[l].startswith('>'):
        pass
    elif l == len(lines)-1:
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        break
    elif lines[l+1].startswith('>'):
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        full_dna=""
    else:
        if full_dna=="":
            full_dna=lines[l]
        else:
            full_dna=full_dna+lines[l]
```

```python
dna_strings
```

```python
motif_matches=[]
for d in range(len(dna_strings)):
    #motif_matches=[]
    if d == len(dna_strings)-1:
        break
    elif d == 0:
        s=SequenceMatcher(None, dna_strings[d], dna_strings[d+1], autojunk=False)
        start_end=s.get_matching_blocks()
        for m in range(len(start_end)):
            m_start=start_end[m][0]
            m_end=start_end[m][0]+start_end[m][2]
            s_motif=dna_strings[d][m_start:m_end]
            motif_matches.append(s_motif)
    else:
        for match in motif_matches:
            if match not in dna_strings[d]:
                motif_matches.remove(match)
```

```python
motif_matches
```

```python
max(motif_matches, key=len)
```
