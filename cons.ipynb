Finds a consensus sequence from a list of FASTA sequences and creates a consensus matrix.

See problem here: https://rosalind.info/problems/cons/


```python
import pandas as pd
```

```python
with open('./Text_Files/rosalind_cons.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n', '')
        
dna_strings=[]
num=0
full_dna=""
for l in range(len(lines)):
    if lines[l].startswith('>'):
        pass
    elif l == len(lines)-1:
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        break
    elif lines[l+1].startswith('>'):
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        full_dna=""
    else:
        if full_dna=="":
            full_dna=lines[l]
        else:
            full_dna=full_dna+lines[l]
```

```python
dna_strings
```

```python
cs_count={}
for n in range(len(dna_strings[0])):
    A,C,G,T=0,0,0,0
    for seq in range(len(dna_strings)):
        nuc = dna_strings[seq][n]
        if nuc == "A": A+=1
        elif nuc == "C": C+=1
        elif nuc == "G": G+=1
        elif nuc == "T": T+=1
        else: break
    cs_count[str(n)]=A,C,G,T
```

```python
cs_count_df=pd.DataFrame(cs_count, index=["A", "C", "G", "T"])
print(str(cs_count_df))
```

```python
cs=""
for col in range(cs_count_df.shape[1]):
    cs_nuc=cs_count_df.loc[cs_count_df.iloc[:,col]==cs_count_df.iloc[:,col].max()].index[0]
    cs=cs+cs_nuc
```
