Finds all signed permuations of a range of numbers (excluding 0)

See associated problem here: https://rosalind.info/problems/sign/

```python
import itertools as it
```

```python
n=4
```

```python
items=range(1,n+1)
def signed_perm(items):
    for p in it.permutations(items):
        for signs in it.product([-1,1], repeat=len(items)):
            yield [a*sign for a,sign in zip(p,signs)]
```

```python
A1=len(list(signed_perm(items)))
```

```python
A2=list(signed_perm(items))
```

```python
print(A1)
for i in A2:
    print(*i)
```
