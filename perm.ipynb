Finds all permutations of a range of numbers.

See associated problem here: https://rosalind.info/problems/perm/

```python
import itertools as it
```

```python
with open('./Text_Files/rosalind_perm.txt', 'r') as f:
    lines = f.readlines()
    n = lines[0].replace('\n','')
    n = int(n)
```

```python
p = list(it.permutations(range(1,n+1)))
```

```python
print(len(p))
print(str(p).replace(' ','').replace('),','\n').replace('(','').replace(',',' ').replace('[','').replace(']','').replace(')',''))
```
