Given: At most 50 DNA strings of approximately equal length, not exceeding 1 kbp, in FASTA format (which represent reads deriving from the same strand of a single linear chromosome). The dataset is guaranteed to satisfy the following condition: there exists a unique way to reconstruct the entire chromosome from these reads by gluing together pairs of reads that overlap by more than half their length. 

Return: A shortest superstring containing all the given strings (thus corresponding to a reconstructed chromosome).

See: https://rosalind.info/problems/long/

```python
# Opens file and reads it
with open('.\\Text_Files\\rosalind_long.txt', 'r') as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n', '')
```

```python
# Combines substrings into strings (formatting from rosalind download) and adds them to a list of fasta sequences
dna_strings=[]
num=0
full_dna=""
for l in range(len(lines)):
    if lines[l].startswith('>'):
        pass
    elif l == len(lines)-1:
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        break
    elif lines[l+1].startswith('>'):
        full_dna=full_dna+lines[l]
        num+=1
        dna_strings.append(full_dna)
        full_dna=""
    else:
        if full_dna=="":
            full_dna=lines[l]
        else:
            full_dna=full_dna+lines[l]
```

```python
#Creates a function that finds the maximum overlap between 2 strings.
def max_overlap(s1, s2):
    max_overlap_len = 0
    max_overlap_str = ""
    max_overlap_dir = 0

    #Check all possible overlaps of s1 ending with s2
    for i in range(1, min(len(s1), len(s2)) + 1):
        if s1[-i:] == s2[:i]:
            if i > max_overlap_len:
                max_overlap_len = i
                max_overlap_str = s1[-i:]
                max_overlap_dir = 0

    #Check all possible overlaps of s2 ending with s1
    for i in range(1, min(len(s1), len(s2)) + 1):
        if s2[-i:] == s1[:i]:
            if i > max_overlap_len:
                max_overlap_len = i
                max_overlap_str = s2[-i:]
                max_overlap_dir = 1

    return[max_overlap_str, max_overlap_len, max_overlap_dir]

max_overlap(dna_strings[0], dna_strings[1])
```

```python
def scaffold_builder(dna_strings):
    scaffold = dna_strings[0] #Start the initial scaffold with the first sequence
    used_strings = [scaffold] #Creates a list of already used strings
    
    while len(used_strings) < len(dna_strings): #Stops once all strings are used
        len_overlap = 0 #Used for indexing later, could probably have just used len()
        overlap = [] # Output of max_overlap(), which is the string, length, and direction of the match
        neighbor = "" #This is used to locally store the string (overlap[0])
        neighbor_index = -1 #Direction of the overlap 
        
        for i, s2 in enumerate(dna_strings):
            if s2 not in used_strings:
                temp_overlap = max_overlap(scaffold, s2)
                if temp_overlap[1] > len_overlap:
                    overlap = temp_overlap
                    len_overlap = temp_overlap[1]
                    neighbor = s2
                    neighbor_index = i
        
        if overlap[2] == 0:
            scaffold = scaffold + neighbor[len_overlap:]
        elif overlap[2] == 1:
            scaffold = neighbor + scaffold[len_overlap:]
        
        used_strings.append(neighbor)
    
    return scaffold
```
